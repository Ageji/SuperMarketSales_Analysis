```python
#Revuene by category
#Filter rows where 'Sale or Return' is 'Sale'
sales_df = df[df['Sale or Return'] == 'sale'].copy()

# Calculate revenue for each row
sales_df['Revenue'] = sales_df['Quantity Sold (kilo)'] * sales_df['Unit Selling Price (RMB/kg)']

# Calculate total revenue by category
total_revenue_by_category = sales_df.groupby('Category Name')['Revenue'].sum()

revenue_df = total_revenue_by_category.reset_index()
revenue_df.columns = ['Category Name', 'Total Revenue']

revenue_df
```

```python
# Plot the total revenue by category
plt.figure(figsize=(10, 6))
plt.bar(revenue_df['Category Name'], revenue_df['Total Revenue'], color='skyblue')

# Add labels and title
plt.xlabel('Category Name')
plt.ylabel('Total Revenue (RMB)')
plt.title('Total Revenue by Category')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Disable scientific notation for y-axis and format it to show large numbers
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.tight_layout()  # Adjust layout to fit labels

plt.show()
```

```python
#  Sales Performance by Hour of the Day
# Extract hour from 'Time' column with a specific format
df['Hour'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.hour

# Filter for sales only
df_sales = df[df['Sale or Return'] == 'sale'].copy()

# Aggregate total sales quantity and revenue by hour
df_sales['Revenue'] = df_sales['Quantity Sold (kilo)'] * df_sales['Unit Selling Price (RMB/kg)']
hourly_sales = df_sales.groupby('Hour').agg({'Quantity Sold (kilo)': 'sum', 'Revenue': 'sum'}).reset_index()

hourly_sales
```

```python
plt.figure(figsize=(12, 6))
# Scatter Plot
sns.scatterplot(x='Hour', y='Revenue', data=hourly_sales, color='purple')
plt.title('Sales Performance by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Reveune')
plt.show()
```

```python
# profit margin per item 
# Calculate Revenue
df['Revenue'] = df['Quantity Sold (kilo)'] * df['Unit Selling Price (RMB/kg)']

#  Calculate Cost considering Loss Rate
df['Cost'] = df['Quantity Sold (kilo)'] * df['Wholesale Price (RMB/kg)']
df['Cost after Loss'] = df['Cost'] * (1 + df['Loss Rate (%)'] / 100)

# Calculate Profit
df['Profit'] = df['Revenue'] - df['Cost after Loss']

# Calculate Profit Margin
df['Profit Margin (%)'] = (df['Profit'] / df['Revenue']) * 100

(df[['Item Name', 'Quantity Sold (kilo)', 'Unit Selling Price (RMB/kg)', 'Revenue', 'Cost after Loss', 'Profit', 'Profit Margin (%)']])
```

```python
# profit per for category
# Calculate Revenue, Cost, and Profit
df['Revenue'] = df['Quantity Sold (kilo)'] * df['Unit Selling Price (RMB/kg)']
df['Cost'] = df['Quantity Sold (kilo)'] * df['Wholesale Price (RMB/kg)']
df['Cost after Loss'] = df['Cost'] * (1 + df['Loss Rate (%)'] / 100)
df['Profit'] = df['Revenue'] - df['Cost after Loss']

# Group by Category Name and calculate total revenue, cost, and profit
category_summary = df.groupby('Category Name').agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Cost=('Cost after Loss', 'sum'),
    Total_Profit=('Profit', 'sum')
)

# Calculate Profit Margin
category_summary['Profit Margin (%)'] = (category_summary['Total_Profit'] / category_summary['Total_Revenue']) * 100

category_summary
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(category_summary.index, category_summary['Profit Margin (%)'], color='skyblue')
plt.xlabel('Category Name')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margin by Category')
plt.xticks(rotation=45)
plt.ylim(0, category_summary['Profit Margin (%)'].max() + 10)  # Adding some space above the highest bar
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Return percentage per category

# Filter rows where 'Sale or Return' is 'Return'
returns_df = df[df['Sale or Return'] == 'return']

# Filter rows where 'Sale or Return' is 'Sale'
sales_df = df[df['Sale or Return'] == 'sale']

# Calculate total and returned quantities for each category
total_sales_by_category = sales_df.groupby('Category Name')['Quantity Sold (kilo)'].sum()
total_returns_by_category = returns_df.groupby('Category Name')['Quantity Sold (kilo)'].sum()

# Handle cases where a category might be in one but not the other
# Fill missing values with 0 for categories with no returns or sales
total_sales_by_category = total_sales_by_category.reindex(total_returns_by_category.index, fill_value=0)
total_returns_by_category = total_returns_by_category.reindex(total_sales_by_category.index, fill_value=0)

# Calculate Return Percentage
return_percentage = (total_returns_by_category / total_sales_by_category) * 100

# Create a DataFrame for results
results_df = pd.DataFrame({
    'Category Name': return_percentage.index,
    'Return Percentage': return_percentage.values
})

# Display formatted Return Percentage for reporting
results_df['Formatted Return Percentage'] = results_df['Return Percentage'].apply(lambda x: f"{x:.2f}%")

(results_df[['Category Name', 'Formatted Return Percentage']])
```

```python
# Convert negative values to positive
# Plotting
plt.figure(figsize=(10, 6))
plt.barh(results_df['Category Name'], results_df['Return Percentage'], color='skyblue')
plt.xlabel('Return Percentage (%)')
plt.ylabel('Category Name')
plt.title('Return Percentage by Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
# Sales Trends Over Time
# Aggregate total quantity sold by year and month
monthly_sales = df.groupby([df['Date'].dt.year, df['Date'].dt.month])['Quantity Sold (kilo)'].sum()

# Convert the multi-level index into a readable format (e.g., '2021-01', '2021-02', etc.)
monthly_sales.index = [f'{int(year)}-{int(month):02d}' for year, month in monthly_sales.index]
monthly_sales_df = monthly_sales.reset_index(name='Total Quantity Sold (kilo)')
monthly_sales_df.columns = ['Month-Year', 'Total Quantity Sold (kilo)']
monthly_sales_df

```

```python
# Plotting
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.xlabel('Year-Month')
plt.ylabel('Total Quantity Sold (kilo)')
plt.title('Sales Trends by Month for Each Year')

# Ensure all months are displayed on the x-axis
plt.xticks(ticks=range(len(monthly_sales.index)), labels=monthly_sales.index, rotation=45)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate revenue loss due to discounts
df['Revenue Loss Due to Discount'] = df.apply(
    lambda row: (row['Unit Selling Price (RMB/kg)'] - row['Wholesale Price (RMB/kg)']) * row['Quantity Sold (kilo)']
    if row['Discount (Yes/No)'] == 'Yes' else 0, axis=1
)

# Total revenue loss due to discounts
total_revenue_loss_due_to_discounts = df['Revenue Loss Due to Discount'].sum()

# Calculate total revenue (without considering discounts)
total_revenue = (df['Unit Selling Price (RMB/kg)'] * df['Quantity Sold (kilo)']).sum()

# Calculate revenue without discounts
revenue_without_discounts = total_revenue - total_revenue_loss_due_to_discounts

revenue_summary = pd.DataFrame({
    'Metric': ['Total Revenue Loss Due to Discounts', 'Total Revenue', 'Revenue Without Discounts'],
    'Amount (RMB)': [total_revenue_loss_due_to_discounts, total_revenue, revenue_without_discounts]
})

revenue_summary
```
